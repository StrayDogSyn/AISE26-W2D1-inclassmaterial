````python
PATH = r'C:\Users\sstucker\Documents\JTC\AISE 25\files\stocks\GOOG.csv'
````

````python
import csv

with open(PATH, 'r') as file:
    reader = csv.DictReader(file)
    for i, row in enumerate(reader):
        print(row)
        if i > 10:
            break
````

````python
data = list(range(65, 256))

with open('unicode.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(['byte', 'character'])
    for i in data:
        # print(i, chr(i))
        writer.writerow([i, chr(i)])
````

````python
import json

FOLDER = r'C:\Users\sstucker\Documents\JTC\AISE 25\files\aircraft'

for file in os.listdir(FOLDER):
    if file.endswith('.json'):
        loc = os.path.join(FOLDER, file)
        with open(loc, 'r') as f:
            try:
                d = json.load(f)
            except json.JSONDecodeError as e:
                print('Error decoding', loc, e)
            print(json.dumps(d, indent=3))
````
